# 3. Feature Enginerring


### 3.1 **Add Attribute Min Discount Per Product**

```python
df['min_discount_per_product'] = df.apply(lambda x : round( (1- x['price_per_unit'] / x['price']) * 100,2) , axis = 1)

df
```

```python
df["min_discount_per_product"] = df["min_discount_per_product"].apply(lambda x: x if x > 0 else 0).astype(str).apply(lambda x: x + "%")

df
```

### 3.2 **Add attribute profit_per_product**


```python
df['profit_per_product'] = df.apply(lambda x : round( (x['price_per_unit'] / x['price'] - 1) * 100,2) , axis = 1)

df
```

```python
df["profit_per_product"] = df["profit_per_product"].apply(lambda x: x if x > 0 else 0).astype(str).apply(lambda x: x + "%")

df
```

### 3.3 Add atribute profit per product (dollar) and total_profit (dollars)

```python
df['profit_per_product (dollar)'] = df.apply(lambda x : round( (x['price_per_unit'] - x['price'])) , axis = 1)

df
```

```python
df["profit_per_product (dollar)"] = df["profit_per_product (dollar)"].apply(lambda x: x if x > 0 else 0)

df
```

```python
df['total_profit (dollars)'] = df.apply(lambda x : round( (x['profit_per_product (dollar)'] * x['quantity_product_sales'])) , axis = 1)

df
```

Since the number of unique values in the 'description' and 'city' columns is quite a lot and it is difficult to check for typo data, here a method will be used to find similar strings within one column.

```python
!pip install rapidfuzz
```

```python
from rapidfuzz import fuzz

# create df customer_city of unique value
unique_description = categorical['description'].unique().tolist()
df_description = pd.DataFrame(unique_description , columns=['unique_description'])

# check similar string
list_desc = []
unique_desc = df_description['unique_description']
for index, row1 in unique_desc.items():
    # skip elements that are already compared
    for row2 in unique_desc.iloc[index+1::]:
        # use a score_cutoff to improve the runtime for bad matches
        score = fuzz.ratio(row1, row2, score_cutoff=99)
        if score:
            list_desc.append([row1, row2, score])

df_match_desc = pd.DataFrame(list_desc, columns=['unique_desc','unique_desc_match','score'])
print(df_match_desc.to_markdown())
```

```python
# create df customer_city of unique value
unique_city = categorical['city'].unique().tolist()
df_city = pd.DataFrame(unique_city , columns=['unique_city'])

# check similar string
list_cty = []
unique_cty = df_city['unique_city']
for index, row1 in unique_cty.items():
    # skip elements that are already compared
    for row2 in unique_cty.iloc[index+1::]:
        # use a score_cutoff to improve the runtime for bad matches
        score = fuzz.ratio(row1, row2, score_cutoff=90)
        if score:
            list_cty.append([row1, row2, score])

df_match_cty = pd.DataFrame(list_cty, columns=['unique_cty','unique_cty_match','score'])
print(df_match_cty.to_markdown())
```
